```python
class minHeap:
    
    def __init__(self,key,value):
        self.key=key
        self.value=value
        
    def __lt__(self,other):
        return self.value<other.value
    
```

```python
import heapq
class Zip:
    
    def __init__(self,path):
        self.__path=path
        self.__fileContent=''
        self.__freqDictionary={}
        self.__heap=[]
        self.__codeBits={}
        self.__reverseCodeBits={}
        self.__encodedText=''
        
    def __readFile(self):
        fileObject = open(self.__path,"r")
        self.__fileContent = fileObject.read()
#         print(fileTxt)
        fileObject.close()
    
    def __makeFreqDictionary(self):
        # this function converts text to freq dictionary
        dict1={}
        for letter in self.__fileContent:
            dict1[letter]=dict1.get(letter,0)+1
        self.__freqDictionary=dict1
        
    def __createMinHeapFromFreqDictionary(self):
        heapList=[]
        
        for key,value in self.__freqDictionary.items():
            minheapObject=minHeap(key,value)
            heapList.append(minheapObject)
        
        heapq.heapify(heapList)
        self.__heap=heapList
        
    def __builTree(self):
    
        while len(self.__heap)>1:
            min_node_1=heapq.heappop(self.__heap)
            min_node_2=heapq.heappop(self.__heap)
            new_node=minHeap(None,min_node_1.value+min_node_2.value)
            new_node.left=min_node_1
            new_node.right=min_node_2
            heapq.heappush(self.__heap,new_node)
            
    def __buildCodesHelper(self,root,code_bits):
        if root is None:
            return
        if root.key is not None:
            self.__codeBits[root.key]=code_bits
            self.__reverseCodeBits[code_bits]=root.key
            return
        self.__buildCodesHelper(root.left,code_bits+'0')
        self.__buildCodesHelper(root.right,code_bits+'1')
    
    def __buildCodes(self):
        root=heapq.heappop(self.__heap)
        self.__buildCodesHelper(root,'')
        
    def __encodedTxt(self):
        encodedText=''
        for letter in self.__fileContent:
            encodedText+=self.__codeBits[letter]
        self.__encodedText=encodedText
        
    def __paddedEncodeTxt(self):
        padded_bit_size=8-(len(self.__encodedText)%8) 
        for i in range(padded_bit_size):
            self.__encodedText+='0'
#         print(padded_bit_size)    
        padded_bit_info='{0:08b}'.format(padded_bit_size)
#         print(padded_bit_info)
        padded_encoded_text=padded_bit_info+self.__encodedText
        return padded_encoded_text
    
    def __getBytesArrayFromEncodedTxt(self,txt):
        arr=[]
        for i in range(0,len(txt),8):
            bytes_str=txt[i:i+8]
            arr.append(int(bytes_str,2))
        return arr    
        
        
    def compress(self):
        ##read the file
        self.__readFile()
#         print(self.__fileContent)
        ## make freq dictionary
        self.__makeFreqDictionary()
#         print(self.__freqDictionary)
        ## make heap
        self.__createMinHeapFromFreqDictionary()
#         print(self.__heap[0].key,self.__heap[0].value,self.__heap[1].key,self.__heap[1].value)
        ## build tree
        self.__builTree()
#         print(len(self.__heap))
        ## make codes
        self.__buildCodes()
#         print(self.__codeBits)
        self.__encodedTxt()
#         print(self.__encodedText)
        padded_encoded_text=self.__paddedEncodeTxt()
        bytes_array=self.__getBytesArrayFromEncodedTxt(padded_encoded_text)
#         print(bytes_array)
        final_bytes=bytes(bytes_array)
#         print(final_bytes)
        ## write this in output file
        output_filepath='output.bin'
        out_file=open(output_filepath,'wb')
        out_file.write(final_bytes)
        out_file.close()
        print('compressed')
        return output_filepath
        
        
        
        
        
        
    
        
        

```
